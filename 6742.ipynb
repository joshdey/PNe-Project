```python
tpf = KeplerTargetPixelFile('secret/kplr100004038-2012179063303_lpd-targ.fits')
```

```python
tpf.interact()
```

```python
tpf.plot()
```

```python
tpf.plot(aperture_mask=tpf.pipeline_mask)
```

```python
lc = tpf.to_lightcurve(aperture_mask=tpf.pipeline_mask)
```

```python
lc = lc.remove_nans().remove_outliers().bin()
```

```python
lc.plot()
```

```python
t = lc.time
y = lc.flux
```

```python
print(len(t),len(y))
```

```python
frequency, power = LombScargle(t,y).autopower(minimum_frequency=0.1, maximum_frequency = 20)
1/frequency[np.argmax(power)]
```

```python
plt.plot(1/frequency, power)
```
